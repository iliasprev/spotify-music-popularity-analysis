```python
df = df.drop(columns=["tidal_popularity"])
```

```python
numeric_cols = [
    "all_time_rank",
    "spotify_streams",
    "spotify_playlist_count",
    "spotify_playlist_reach",
    "youtube_views",
    "youtube_likes",
    "tiktok_posts",
    "tiktok_likes",
    "tiktok_views",
    "youtube_playlist_reach",
    "apple_music_playlist_count",
    "airplay_spins",
    "siriusxm_spins",
    "deezer_playlist_count",
    "deezer_playlist_reach",
    "amazon_playlist_count",
    "pandora_streams",
    "pandora_track_stations",
    "soundcloud_streams",
    "shazam_counts"
]
```

```python
for col in numeric_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(",", "", regex=False)
    )
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df[numeric_cols].info()
```

```python
df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce")

df["release_date"].info()
```

## Missing Values Strategy

- Missing values represent unavailable platform data
- Values were not imputed to avoid introducing bias
- Nulls are preserved for transparency and accurate analysis
